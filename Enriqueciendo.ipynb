```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data/clean_compras.csv')
```

```python
data['FECHA'] = pd.to_datetime(data['FECHA'])
```

```python
data['DIA'] = data['FECHA'].apply(lambda x: x.day)
data['MES'] = data['FECHA'].apply(lambda x: x.month)
data['YEAR'] = data['FECHA'].apply(lambda x: x.year)
```

```python
data.dtypes
```

```python
ben_sales = data['BENEFICIARIO'].value_counts().to_dict()
```

```python
list(ben_sales.keys())[0:5]
```

```python
data['BEN_SALES'] = data['BENEFICIARIO'].apply(lambda x: ben_sales[x])
```

```python
sheet_mean_sales = data.groupby('SHEET')['IMPORTE'].mean()
```

```python
sheet_mean_sales
```

```python
data['GASTO PROMEDIO'] = data['SHEET'].apply(lambda x: sheet_mean_sales[x])
```

```python
pd.DataFrame(data['BENEFICIARIO'].unique()).to_csv('./data/beneficiarios.csv', index=None)
```

```python
linkedin_mty = pd.read_csv('./data/linkedin_mty.csv')
```

```python
has_site = {}
for i, ele in linkedin_mty.fillna(True).iterrows():
  has_site[ele['query']] = True if ele['error'] == True else False
```

```python
len(has_site.keys())
```

```python
data['BENEFICIARIO'].unique().size
```

```python
data['HAS_LINKEDIN'] = data['BENEFICIARIO'].apply(lambda x: has_site[x.replace('  ', ' ')])
```
